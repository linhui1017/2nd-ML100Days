```python
# 載入基礎套件
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### 弱相關

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

# 呼叫 numpy 裡的相關矩陣函數 (corrcoef)
np.corrcoef(x, y)
```

```python
# c
v = np.vstack((x,y))
np.cov(v, bias =True)/ (np.std(x) * np.std(y))
```

```python
# 將分布畫出來看看吧
plt.scatter(x, y)
```

### 正相關

```python
# 隨機生成 1000 個介於 0~50 的數 x
x = np.random.randint(0, 50, 1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊
y = x + np.random.normal(0, 10, 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x, y)
```

```python
# c
v = np.vstack((x,y))
np.cov(v, bias =True) / (np.std(x) * np.std(y))
```

```python
# 再看看正相關的 x,y 分布
plt.scatter(x, y)
```
